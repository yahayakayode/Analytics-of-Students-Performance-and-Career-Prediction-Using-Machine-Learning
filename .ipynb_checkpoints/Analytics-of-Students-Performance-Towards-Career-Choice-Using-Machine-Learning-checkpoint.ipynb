It can be observed that the columns are not uniformly names as some are in capital letters while some in small letters, hence the need to formalise the clumns for a better analysis and prediction.

```python
std_career_df.columns = std_career_df.columns.str.replace(' ', '_').str.replace('/', '_').str.replace('?', '').str.lower()
```

```python

cols_with_obj_type = [c for c,d in zip(std_career_df.columns,std_career_df.dtypes) if d == np.object]
```

```python
le = LabelEncoder()
```

```python
for col in cols_with_obj_type:
    std_career_df[col] = le.fit_transform(std_career_df[col])

print(col)
```

```python

X = std_career_df.drop(['suggested_job_role'],axis=1)
y = std_career_df.suggested_job_role
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X , y , test_size = 0.3 , random_state = 1)
```

```python

#Logistic Regression - 
logreg = LogisticRegression()

logreg.fit(X_train , y_train)

y_pred = logreg.predict(X_test)

print(logreg.score(X_test , y_test))
```

```python
# Decision Tree Classifier
from sklearn.tree import DecisionTreeClassifier 

dt = DecisionTreeClassifier(criterion = 'gini' , max_depth = 3)

dt = dt.fit(X_train , y_train)
y_pred = dt.predict(X_test)

print("Accuracy", metrics.accuracy_score(y_test , y_pred))
```
